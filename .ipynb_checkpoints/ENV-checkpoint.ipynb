```python
class Env():
    def __init__(self):
        self.Cam=Cam.Cam()
    
    def get_raw_frames(self,time=3):
        return self.Cam.get_state(Cam.fishcam,time)
        
    
```

```python
ep=1
state_num=0
env=Env()
frames = env.get_raw_frames()
for i in frames:
    frame=Image.fromarray(i)
    frame.save('RL_DATA/ep%d/%d'%(ep,state_num)
```
